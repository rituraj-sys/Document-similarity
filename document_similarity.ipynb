```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embed = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")
```

```python
document = ["Narendra modi is the prime minister of India, and he is one of the greate leaders in the world",
            "Vladimir Putin is te president of Russia and also one of the nuclear power on earth",
            "Donald Trump is the good leader in USA",
            "Elon Musk is the ceo and founder of Tesla and a crazy leader like nuclear fusion."]
query = "Who is the prime minister of India?"
```

```python
doc_embed = embed.embed_documents(document)
query_embed = embed.embed_query(query)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity([query_embed], doc_embed)[0]
print(similarity)
```

```python
ind_sim = similarity.argmax()
print(ind_sim)
print(query)
print(document[ind_sim])
```
